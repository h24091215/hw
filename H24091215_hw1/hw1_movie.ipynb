```python
#第四題
import pandas as pd
df=pd.read_csv('IMDB-Movie-Data.csv')
df.columns
df['Actors']=df['Actors'].str.split('|')

list1=[]#directors
for i in df['Director']:
    if i not in list1:
        list1.append(i)

dict1={}
c=0
d=0
e=0
for i in list1:
    s=set([])
    b=df[df['Director'].str.contains(i)]
    for j in range(len(b)):
        s=s|set(b.iloc[j,4])
    
    a=set([])
    for j in s:
        a.add(j.strip())
    dict1[i]=len(a)
    
for i in list1:
    if dict1[i]>c:
        c=dict1[i]
        first=i
    elif dict1[i]==c and i!=first:
        print('first:',i)
    elif dict1[i]>d:
        d=dict1[i]
        second=i
    elif dict1[i]==d and i!=second:
        print('second:',i)
    elif dict1[i]>e:
        e=dict1[i]
        third=i
    elif dict1[i]==e and i!=third:
        print('third:',i)
print('first:',first,'\nsecond:',second,'\nthird:',third)
```

```python
#第五題
import pandas as pd
df=pd.read_csv('IMDB-Movie-Data.csv')
df.columns
df['Actors']=df['Actors'].str.split('|')
df['Genre']=df['Genre'].str.split('|')

df2=pd.DataFrame(columns=['Actors','Genre'])

for i in range(len(df)):
    for j in df['Actors'][i]:
        for k in df['Genre'][i]:
            df2=df2.append({'Actors':j.strip(),'Genre':k.strip()},ignore_index=True)

list1=[]#actors
b=0
c=0
for i in df['Actors']:
    for j in i:
        if j not in list1:
            list1.append(j.strip())
for i in list1:
    df3=df2.groupby('Actors').get_group(i)
    #df3['count']=df3['Genre'].count()
    #print(df3)
    s=set()
    for j in df3['Genre']:
        #s=set()
        J=[]
        J.append(j)
        s=s|set(J)
    #print(s)
    if len(s)>b:
        b=len(s)
        first=i
    elif len(s)==b and i!=first:
        print('first',i)
    
print('first',first)
```

```python
#第六題
import pandas as pd
df=pd.read_csv('IMDB-Movie-Data.csv')
df.columns
df['Actors']=df['Actors'].str.split('|')

df2=pd.DataFrame(columns=['Actors','Year'])

for i in range(len(df)):
    for j in df['Actors'][i]:
        #print(i,j)
        df2=df2.append({'Actors':j.strip(),'Year':df['Year'][i]},ignore_index=True)
#print(df2)
df3=df2.groupby('Actors').aggregate([pd.DataFrame.max,pd.DataFrame.min])
df3['gap']=df3["Year"]["max"]-df3["Year"]["min"]
df3=df3.sort_values(by=['gap'],ascending=False)
df3['gap'][0]
print(df3["gap"][df3["gap"]==df3["gap"][0]])
```
